```python
#Q3Create a Pandas DataFrame that contains the following data:
data = {'Name': ['Alice', 'Bob', 'Clarie'],'Age': [25, 30, 27], 'Gender': ['Female', 'Male', 'Female']}
df = pd.DataFrame(data)
print(df)
```

```python
#Q4  A DataFrame in pandas is a 2-dimensional labelled data structure whereas Series is a 1-dimensional labelled array that may hold any kind of data.
import pandas as pd

# create a Series
ser = pd.Series([1, 2, 3, 4, 5], name='numbers')

# create a DataFrame with two columns
df = pd.DataFrame({'name': ['Alice', 'Bob', 'Clarie', 'Angel', 'Dev'],
                   'age': [25, 30, 35, 40, 45]})

# print the Series and DataFrame
print(ser)
print(df)
```
